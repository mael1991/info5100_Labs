```python
#  WINE QUALITY DATASET VISUALIZATION NOTEBOOK

# --- Step 1: Import libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inline
%matplotlib inline

# --- Step 2: Load dataset ---
# You can download it from: https://archive.ics.uci.edu/ml/datasets/Wine+Quality
# We'll use the red wine data version:
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, sep=';')

# --- Step 3: Preview the dataset ---
print("Dataset shape:", df.shape)
df.head()

# --- Step 4: Basic statistics ---
print("\nSummary statistics:")
print(df.describe())

# --- Visualization 1: Histogram / Distribution Plot ---
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='alcohol', kde=True, color='purple')
plt.title('Distribution of Alcohol Content in Wine', fontsize=14)
plt.xlabel('Alcohol (%)')
plt.ylabel('Frequency')
plt.show()

# --- Visualization 2: Correlation Heatmap ---
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Wine Features', fontsize=14)
plt.show()

# --- Visualization 3: Boxplot (Alcohol vs Quality) ---
plt.figure(figsize=(8,5))
sns.boxplot(x='quality', y='alcohol', data=df, palette='Set2')
plt.title('Alcohol Content Across Wine Quality Levels', fontsize=14)
plt.xlabel('Quality Rating')
plt.ylabel('Alcohol (%)')
plt.show()

# --- Visualization 4: Scatter Plot with Regression Line ---
plt.figure(figsize=(8,6))
sns.regplot(x='alcohol', y='density', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Relationship Between Alcohol and Density', fontsize=14)
plt.xlabel('Alcohol (%)')
plt.ylabel('Density')
plt.show()

# --- Step 5: Optional insights ---
print("\nAverage wine quality:", round(df['quality'].mean(), 2))
print("Correlation between alcohol and quality:", round(df['alcohol'].corr(df['quality']), 3))
```
